## OpenCV를 사용하여 사진속에 왈도를 찾아내기



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 이미지 불러오기
image1 = cv2.imread('images/beach_waldo.jpg')
image2 = cv2.imread('images/waldo.jpg')


plt.figure(figsize=(12,5))
plt.subplot(1,2,1) # 바닷가 사진 출력하기
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))

plt.subplot(1,2,2) # 왈도 사진 출력하기
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np


# 흑백으로 전환
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# 두개의 흑백이미지를 비교 및 매칭된 부위 위치정보
result = cv2.matchTemplate(gray1, gray2, cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result) #max_loc 은 좌상단이다


# 위치정보를 통해 빨간색 직사격형 출력 
top_left = max_loc
bottom_right = (top_left[0] + 50, top_left[1] + 50)
cv2.rectangle(image1, top_left, bottom_right, (0,0,255), 3)

cv2.imshow('Where is Waldo?', image1)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
