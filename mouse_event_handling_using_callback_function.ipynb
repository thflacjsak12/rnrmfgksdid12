```python
import cv2
import numpy as np
# 검은색 바탕화면의 윈도우만들기

win_title = 'Mouse Event'
image = np.zeros((512,512,3), np.uint8)
cv2.imshow(win_title, image)

def clickMouse(event, x, y, flags, param):
    print(event, x, y)
    if event == cv2.EVENT_LBUTTONDOWN: # 마우스 왼쪽버튼이 눌림
        cv2.circle(image, (x,y), 30, (255,255,255), -1)
        cv2.imshow(win_title, image)
        
    if event == cv2.EVENT_RBUTTONDOWN: # 마우스 오른쪽버튼이 눌림
        cv2.circle(image, (x,y), 30, (255,255,255), -1)
        cv2.rectangle(image, (x-10, y-10), (x+10, y+10), (255,0,0), -1)
        cv2.imshow(win_title, image)
        
cv2.setMouseCallback(win_title, clickMouse) # 마우스 callback 함수를 GUI 윈도우에 등록

while True:
    if cv2.waitKey(0) & 0xFF == 27: # ESC버튼을 누르면 종료
        break;

cv2.destroyAllWindows()
```
